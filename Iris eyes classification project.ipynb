```python
l=[]
target=[]
for i in range(1,41,1):
    for j in range(1,6,1):
        img=plt.imread("F://iris//U"+str(i)+"//"+str(j)+".jpg")
        new=img.mean(axis=1)
        target.append(i)
        l.append(new)
```

```python
l
```

```python
df=pd.DataFrame(l)
```

```python
df
```

```python
target
```

```python
df['Target']=target
```

```python
df
```

```python
test=df.iloc[4::5]
```

```python
train=df.drop(np.arange(4,200,5))
```

```python
X_train=train.iloc[:,:-1]
```

```python
X_test=test.iloc[:,:-1]
```

```python
y_train=train['Target']
```

```python
y_test=test['Target']
```

```python
from sklearn import svm
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(model.predict(X_test),y_test)
```
