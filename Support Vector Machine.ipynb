```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
import sklearn.datasets as data
```

```python
data.load_wine()
```

```python
feature_data=data.load_iris().data
feature_data
```

```python
target=data.load_iris().target
target
```

```python
target_name=data.load_iris().target_names
target_name
```

```python
feature_name=data.load_iris().feature_names
feature_name
```

```python
data=pd.DataFrame(feature_data,columns=feature_name)
data['target']=target
```

```python
data
```

```python
data['feature_names']=pd.Categorical.from_codes(target,feature_name)
```

```python
data
```

```python
help(train_test_split)
```

```python
X=data.iloc[:,0:4]
y=data.iloc[:,4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = svm.SVC(kernel='poly')
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred,y_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(pred,y_test)
```

```python
data
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#parameters model , Independent variables , Dependent Variable , cv=How many times we wish to cross validate
cross_val_score(model,X,y,cv=4)
```
